```python
import torch
```

```python
def fix_torch_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

fix_torch_seed()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("upstage/TinySolar-248m-4k")
model = AutoModelForCausalLM.from_pretrained("upstage/TinySolar-248m-4k")
```

```python
prompt = "I am an engineer. I love"
```

```python
inputs = tokenizer(prompt, return_tensors="pt")
```

```python
from transformers import TextStreamer
streamer = TextStreamer(
    tokenizer,
    skip_prompt=True, # If you set to false, the model will first return the prompt and then the generated text
    skip_special_tokens=True
)
```

```python
outputs = model.generate(
    **inputs,
    streamer=streamer,
    use_cache=True,
    max_new_tokens=128,
    do_sample=False,
    temperature=0.0,
    repetition_penalty=1.1
)
```

```python
def find_max(numbers):
   max = 0
   for num in numbers:
       if num > max:
           max = num
   return max
```
